Recreating the Gapminder world develoment chart with Python using data from 2002

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
gapminder = pd.read_csv ("gapminder_full 2002.csv")
```

```python
gapminder.head()
```

```python
plt.scatter (gapminder.gdp_cap, gapminder.life_exp)
```

```python
plt.scatter (gapminder.gdp_cap, gapminder.life_exp)
plt.xscale('log')
```

```python
plt.scatter (gapminder.gdp_cap, gapminder.life_exp)
plt.xscale('log')
plt.title("World development in 2002")
plt.xlabel("GDP per capita [in USD]")
plt.ylabel ("Life expectancy")
```

```python
plt.scatter (gapminder.gdp_cap, gapminder.life_exp)
plt.xscale('log')
plt.title("World development in 2002")
plt.xlabel("GDP per capita [in USD]")
plt.ylabel ("Life expectancy")
plt.xticks([1000, 10000, 100000],
         ["1k", "10k", "100k"])
```

```python
np_pop = np.array(gapminder.population)
np_pop = np_pop * 2
```

```python
sns.scatterplot (gapminder.gdp_cap, gapminder.life_exp, hue=gapminder.continent, size=np_pop)
plt.xscale('log')
plt.grid('True')
plt.title("World development in 2002")
plt.xlabel("GDP per capita [in USD]")
plt.ylabel ("Life expectancy")
plt.xticks([1000, 10000, 100000],
         ["1k", "10k", "100k"])
```

```python
plt.figure(dpi=150)
sns.scatterplot (gapminder.gdp_cap, gapminder.life_exp, hue=gapminder.continent, size=np_pop)
plt.xscale('log')
plt.grid('True')
plt.title("World development in 2002")
plt.xlabel("GDP per capita [in USD]")
plt.ylabel ("Life expectancy")
plt.xticks([1000, 10000, 100000],
         ["1k", "10k", "100k"])
```

```python
plt.figure(dpi=150)
sns.scatterplot (gapminder.gdp_cap, gapminder.life_exp, hue=gapminder.continent, size=np_pop, sizes=(20,400), legend=False)
plt.xscale('log')
plt.grid('True')
plt.title("World development in 2002")
plt.xlabel("GDP per capita [in USD]")
plt.ylabel ("Life expectancy")
plt.xticks([1000, 10000, 100000],
         ["1k", "10k", "100k"])
```

```python
plt.figure(dpi=150)
sns.scatterplot (gapminder.gdp_cap, gapminder.life_exp, hue=gapminder.continent, size=np_pop, sizes=(20,400), legend=False)
plt.xscale('log')
plt.grid('True')
plt.title("World development in 2002", fontsize=20)
plt.xlabel("GDP per capita [in USD]")
plt.ylabel ("Life expectancy [in years]")
plt.xticks([1000, 10000, 100000],
         ["1k", "10k", "100k"])
```

```python
plt.figure(dpi=150)
sns.scatterplot (gapminder.gdp_cap, gapminder.life_exp, hue=gapminder.continent, size=np_pop, sizes=(20,400), legend=False)
plt.xscale('log')
plt.grid('True')
plt.title("World development in 2002", fontsize=20)
plt.xlabel("GDP per capita [in USD]")
plt.ylabel ("Life expectancy [in years]")
plt.xticks([1000, 10000, 100000],
         ["1k", "10k", "100k"])
plt.text(3119.280896, 72.028, "China")
plt.text(1746.769454, 62.879, "India", fontsize=8)
plt.text(39097.09955, 77.31, "United States")
plt.text(2873.91287, 68.588, "Indonesia")
plt.text(8131.212843, 71.006, "Brazil")
plt.text(2092.712441, 63.61, "Pakistan", fontsize=8)
plt.text(1136.39043, 62.013, "Bangladesh", fontsize=8)
plt.text(28604.5919, 82, "Japan")
plt.text(1615.286395, 46.608, "Nigeria")
plt.text(10742.44053, 74.902, "Mexico")
```

NOTES: The different colours represent different Continents and the size of each scatter point is the population of that country. So I've only named the ten most populous nations (the largest circles on the graph).
